```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime

# Function to get the opening price of November and the closing price of December
def get_nov_dec_prices(ticker, year):
    # Download data for November and December
    nov_data = yf.download(ticker, start=f"{year}-11-01", end=f"{year}-11-05")
    dec_data = yf.download(ticker, start=f"{year}-12-27", end=f"{year+1}-01-01")

    # Get the first valid opening price in November
    nov_open_price = nov_data['Open'].dropna().iloc[0] if not nov_data.empty else None

    # Get the last valid closing price in December
    dec_close_price = dec_data['Close'].dropna().iloc[-1] if not dec_data.empty else None

    return nov_open_price, dec_close_price

# Process data for the given range of years
data = []
for year in range(2014, 2024):
    nov_open, dec_close = get_nov_dec_prices('NKE', year)
    if nov_open is not None and dec_close is not None:
        data.append({'Year': year, 'Nov Open': nov_open, 'Dec Close': dec_close})

df = pd.DataFrame(data)
df.set_index('Year', inplace=True)

# Plotting the data
fig, ax = plt.subplots(figsize=(10, 6))

# Plot line segments from Nov Open to Dec Close for each year
for year in range(2014, 2023):
    ax.plot([datetime.date(year, 11, 1), datetime.date(year, 12, 31)], 
            [df.loc[year, 'Nov Open'], df.loc[year, 'Dec Close']], 
            marker='o', label=f'{year}')

ax.set_title('Nike Stock Price Movement: November Open to December Close (2014-2023)')
ax.set_xlabel('Date')
ax.set_ylabel('Price (USD)')
ax.legend()
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.show()


```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Function to get the opening price of November and the closing price of December
def get_nov_dec_prices(ticker, year):
    nov_data = yf.download(ticker, start=f"{year}-11-01", end=f"{year}-11-05")
    dec_data = yf.download(ticker, start=f"{year}-12-27", end=f"{year+1}-01-01")

    nov_open_price = nov_data['Open'].dropna().iloc[0] if not nov_data.empty else None
    dec_close_price = dec_data['Close'].dropna().iloc[-1] if not dec_data.empty else None

    return nov_open_price, dec_close_price

# Calculate the portfolio value over the years
initial_investment = 4600  # Initial investment amount in USD
leverage = 2.5  # Leverage factor
shares = 100  # Number of shares to trade
portfolio_value = initial_investment
years = []
values = []

for year in range(2014, 2024):
    if year == 2016:  # Skip the year 2016
        continue
    
    nov_open, dec_close = get_nov_dec_prices('NKE', year)
    if nov_open is not None and dec_close is not None:
        # Calculate investment value
        investment = portfolio_value 
        buy_cost = nov_open * shares
        sell_revenue = dec_close * shares

    
        profit_loss = (sell_revenue - buy_cost) 
        
        if profit_loss > 0:
            profit_loss *= leverage
        else:
            profit_loss *= leverage
            
        portfolio_value += profit_loss / leverage

        years.append(year)
        values.append(portfolio_value)

# Plotting the data
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(years, values, marker='o', linestyle='-')

ax.set_title('Portfolio Value Over Time with Nike Stock Strategy (100 shares, 2.5x leverage)')
ax.set_xlabel('Year')
ax.set_ylabel('Portfolio Value (USD)')
plt.xticks(years)  # Set x-axis ticks to be each year
plt.grid(True)
plt.show()
```
